```python
# Import libraries
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("All libraries imported successfully")
```

```python
# Pull active institution data from FDIC API
url = "https://banks.data.fdic.gov/api/institutions"

params = {
    "filters": "ACTIVE:1",
    "fields": "CERT,NAME,CITY,STNAME,ASSET,REPDTE",
    "limit": 10000,
    "offset": 0,
    "output": "json"
}

response = requests.get(url, params=params)
data = response.json()

# Convert to dataframe
institutions_df = pd.DataFrame([item["data"] for item in data["data"]])

print(f"Total active banks pulled: {len(institutions_df)}")
print(f"Columns: {list(institutions_df.columns)}")
institutions_df.head()
```

```python
# Pull failed bank list from FDIC API
failures_url = "https://banks.data.fdic.gov/api/failures"

failure_params = {
    "fields": "CERT,NAME,FAILDATE,SAVR,RESTYPE",
    "limit": 10000,
    "offset": 0,
    "output": "json"
}

response_failures = requests.get(failures_url, params=failure_params)
failures_data = response_failures.json()

# Convert to dataframe
failures_df = pd.DataFrame([item["data"] for item in failures_data["data"]])

print(f"Total failed banks: {len(failures_df)}")
print(f"Columns: {list(failures_df.columns)}")
failures_df.head()
```
